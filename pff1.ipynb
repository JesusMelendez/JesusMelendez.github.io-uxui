```python
def cal_cum_returns(df, col):
  ser = df[col]
  return (ser
          .sub(ser.iloc[0])
          .div(ser.iloc[0])
          )
```

#### On-balance Volume (OBV)

Indicator of momentum using volume

$$
OBV = OBV_{\text{prev}} +
\begin{cases}
\text{volume}, & \text{if } \text{close} > \text{close}_{\text{prev}} \\
0, & \text{if } \text{close} = \text{close}_{\text{prev}}\\
\text{-volume}, & \text{if } \text{close} < \text{close}_{\text{prev}}
\end{cases}
\\where:\\
OBV = \text{Currente on-balance volume level}\\
OBV_{\text{prev}} = \text{Previous on-balnce voluymen level}\\
volume = \text{Latest trading volumen amount}
$$





```python
aapl.Close.head(2)
```

```python
aapl.Close.shift(1).head(2)
```

```python
# naive
def calc_obv(df):
  df = df.copy()
  df['OBV'] = 0.0
  #loop throught the data and calculate OBV
  for i in range(1,len(df)):
    if df['Close'][i] > df['Close'][i-1]:
      df['OBV'][i] = df['Volume'][i] + df['OBV'][i-1]
    elif df['Close'][i] < df['Close'][i-1]:
      df['OBV'][i] = df['OBV'][i-1] - df['Volume'][i]
    else:
      df['OBV'][i] = df['OBV'][i-1]
  return df
```

```python
def calc_obv(df,close_col='Close',volume_col='Volume'):
  close = df[close_col]
  vol = df[volume_col]
  close_shift = close.shift(1)
  return (df
          .assign(vol = np.select(
                                  condlist = [close > close_shift,
                                              close == close_shift,
                                              close < close_shift],
                                  choicelist = [vol,0,-vol]
                                  ),
                  obv = lambda df_: df_.vol.fillna(0).cumsum()
                  )
          ['obv']
          )

```

```python
(aapl
 .assign(obv=calc_obv(aapl)))
```

#### Acumulation | Distribution Indicator (A/D)

Goals

- Show more complicated example
 $$
 MFM = \frac{(Close-Low ) -(High - Close)}{High-Low}
 $$
 **where:**

 MFM = Money Flow Multiplier\
 Close = Closing price\
 Low = Low price for the period\
 High = High price for the period

 Money Flow Volume = MFM x Period Volume\
 A/D = Previous A/D + CMFV\
 **where:**\
CMFV = Current period money flow volume




```python
(aapl
 .assign(mfm =((aapl.Close - aapl.Low)-(aapl.High - aapl.Close))/(aapl.High - aapl.Low),
         mfv = lambda df_:df_.mfm * df_.Volume,
         cmfv= lambda df_:df_.mfv.cumsum()
         )
)
```

```python
def calc_ad(df,close_col='Close',high_col='High',low_col='Low',volume_col='Volume'):
  close = df[close_col]
  high = df[high_col]
  low = df[low_col]
  vol = df[volume_col]
  return (df
          .assign(mfm =((close - low)-(high - close))/(high - low),
                  mfv = lambda df_:df_.mfm * df_.Volume,
                  cmfv= lambda df_:df_.mfv.cumsum()
                  )
          .cmfv
          )
```

```python
(aapl
 .assign(ad=calc_ad)
 .ad.plot(figsize=(10,3))
 )
```

Relative Strength Index is a popular momentum indicator.

- \> 70 overbought

$$

$$

Where RS is the relative strength, calculated as:

$$
RS = \frac{AverageGain}{AverageLoss}

- Average of first 14 values for first value
